```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from tkinter import *

# Function for clearing the contents of all entry boxes and text area.
def clearAll():
    negativeField.delete(0, END)
    neutralField.delete(0, END)
    positiveField.delete(0, END)
    overallField.delete(0, END)
    textArea.delete(1.0, END)

# Function to print sentiments of the sentence.
def detect_sentiment():
    # Get the input content from text box
    sentence = textArea.get("1.0", "end").strip()

    # Create a SentimentIntensityAnalyzer object
    sid_obj = SentimentIntensityAnalyzer()

    # Get the sentiment dictionary from the polarity_scores method
    sentiment_dict = sid_obj.polarity_scores(sentence)

    # Insert the sentiment scores into their respective fields
    negativeField.insert(10, f"{sentiment_dict['neg']*100:.2f}% Negative")
    neutralField.insert(10, f"{sentiment_dict['neu']*100:.2f}% Neutral")
    positiveField.insert(10, f"{sentiment_dict['pos']*100:.2f}% Positive")

    # Determine overall sentiment
    if sentiment_dict['compound'] >= 0.05:
        overall_sentiment = "Positive"
    elif sentiment_dict['compound'] <= -0.05:
        overall_sentiment = "Negative"
    else:
        overall_sentiment = "Neutral"
    overallField.insert(10, overall_sentiment)

# Driver Code
if __name__ == "__main__":
    # Create a GUI window
    gui = Tk()

    # Set the background color of GUI window
    gui.config(background = "light green")

    # Set the name and configuration of the tkinter GUI window
    gui.title("Sentiment Detector")
    gui.geometry("400x400")

    # Create and place the widgets in the GUI window
    Label(gui, text="Enter Your Sentence", bg="light green").grid(row=0, column=1, padx=10, pady=10)
    textArea = Text(gui, height=5, width=40, font="lucida 13")
    textArea.grid(row=1, column=1, padx=10, pady=10)

    Button(gui, text="Check Sentiment", fg="Black", bg="Red", command=detect_sentiment).grid(row=2, column=1, pady=10)

    Label(gui, text="Negative Sentiment:", bg="light green").grid(row=3, column=0, padx=10, pady=10)
    negativeField = Entry(gui)
    negativeField.grid(row=3, column=1, padx=10, pady=10)

    Label(gui, text="Neutral Sentiment:", bg="light green").grid(row=4, column=0, padx=10, pady=10)
    neutralField = Entry(gui)
    neutralField.grid(row=4, column=1, padx=10, pady=10)

    Label(gui, text="Positive Sentiment:", bg="light green").grid(row=5, column=0, padx=10, pady=10)
    positiveField = Entry(gui)
    positiveField.grid(row=5, column=1, padx=10, pady=10)

    Label(gui, text="Overall Sentiment:", bg="light green").grid(row=6, column=0, padx=10, pady=10)
    overallField = Entry(gui)
    overallField.grid(row=6, column=1, padx=10, pady=10)

    Button(gui, text="Clear", fg="Black", bg="Red", command=clearAll).grid(row=7, column=0, padx=10, pady=10)
    Button(gui, text="Exit", fg="Black", bg="Red", command=exit).grid(row=7, column=1, padx=10, pady=10)

    # Start the GUI
    gui.mainloop()
```
